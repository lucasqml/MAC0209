```python
def atualizaEscala(pos, tmp):
    escalaP = -0.2
    escalaT = 0.1
    
    for i in range(0, len(pos)):
        pos[i] *= escalaP
        tmp[i] *= escalaT

atualizaEscala(posicao1, tempo1)
atualizaEscala(posicao2, tempo2)
atualizaEscala(posicao3, tempo3)
atualizaEscala(posicao4, tempo4)
atualizaEscala(posicao5, tempo5)
```

```python
def calculaVel(posicao, tempo):
    velocidade = [0]
    
    for i in range(1, len(posicao)):
        delta_s = posicao[i]-posicao[i-1]
        delta_t = tempo[i]-tempo[i-1]
        velocidade.append(delta_s/delta_t)
    return velocidade

velocidade1 = calculaVel(posicao1, tempo1)
velocidade2 = calculaVel(posicao2, tempo2)
velocidade3 = calculaVel(posicao3, tempo3)
velocidade4 = calculaVel(posicao4, tempo4)
velocidade5 = calculaVel(posicao5, tempo5)
```

```python
import matplotlib.pyplot as plt

# CONSTANTES
GRAVITY = 980
k = 0.02

def aceleration():
    return -GRAVITY

def acelerationResistence(vt):
    return -GRAVITY - k*vt*abs(vt)

def velocity(vt, delta_t): # v(t + delta_t)  = a(t)*delta_t + v(t)
    return aceleration()*delta_t + vt

def velocityResistence(vt, delta_t): # v_r(t + delta_t)  = a_r(t)*delta_t + v_r(t)
    return acelerationResistence(vt)*delta_t + vt 

def position(vt, st, delta_t): # s(t + delta_t) = v(t)*delta_t + s(t)
    return vt*delta_t + st 

def attVector(s0, times):
    sf = s0
    vf = 0
    posEuler = []
    velEuler = []
    posEuler.append(s0)
    velEuler.append(vf)
    i = 0
    while (i < len(times) - 1):
        delta_t = times[i+1] - times[i]
        vf = velocity(vf, delta_t)
        sf = position(vf, sf, delta_t)
        posEuler.append(sf)
        velEuler.append(vf)
        i += 1

    return (posEuler, velEuler)

def attVectorResistence(s0, times):
    sf = s0
    vf = 0
    posEuler = []
    velEuler = []
    posEuler.append(s0)
    velEuler.append(vf)
    i = 0
    while (i < len(times) - 1):
        delta_t = times[i+1] - times[i]
        vf = velocityResistence(vf, delta_t)
        sf = position(vf, sf, delta_t)
        posEuler.append(sf)
        velEuler.append(vf)
        i += 1
    
    return (posEuler, velEuler)
```

```python
def plot_freefall(distancias, times, i):
    posEuler, velEuler = attVector(distancias[0], times)
    posEulerResistencia, velEulerResistencia = attVectorResistence(distancias[0], times)
    
    fig, ax = plt.subplots(1, figsize=(10,5))
    ax.plot(times, posEuler, label='Método de Euler')
    ax.plot(times, posEulerResistencia, label='Método de Euler com resistencia')
    ax.plot(times, posicao1, label='Distâncias reais', linestyle='', marker='X', color='red')
    ax.legend()
    ax.set_title('Gráficos da filmagem ' + i, fontsize=16)
    ax.set_xlabel('Tempo decorrido (s)', fontsize=14);
    ax.set_ylabel('Distância percorrida (m)', fontsize=14);
    
    fig, ax = plt.subplots(1, figsize=(10,5))
    ax.plot(times, velEuler, label='Método de Euler')
    ax.plot(times, velEulerResistencia, label='Método de Euler com resistencia')
    ax.plot(times, velocidade1, label='Velocidades reais',linestyle='', marker='X', color='red')
    ax.legend()
    ax.set_title('Gráficos da filmagem ' + i, fontsize=16)
    ax.set_xlabel('Tempo decorrido (s)', fontsize=14);
    ax.set_ylabel('Velocidade no trecho (m/s)', fontsize=14);
    
    print("Posição sem resistencia "+ i + ":")
    print(posEuler)
    print("Posição com resistencia "+ i + ":")
    print(posEulerResistencia)
    print("Velocidade sem resistencia "+ i + ":")
    print(velEuler)
    print("Velocidade com resistencia "+ i + ":")
    print(velEulerResistencia)
```

```python
plot_freefall(posicao1, tempo1, '1')

plot_freefall(posicao2, tempo2, '2')

plot_freefall(posicao3, tempo3, '3')

plot_freefall(posicao4, tempo4, '4')

plot_freefall(posicao5, tempo5, '5')

```
